```python
from bs4 import BeautifulSoup
import requests
import re
```

```python
def getData(link):
  r = requests.get(link)
  soup = BeautifulSoup(r.content, "html5lib")
  soup.prettify()
  comments = soup.find_all("img", alt=re.compile("Portrait.png"))
  return comments
```

```python
def getImages():
    store = []
    r = requests.get("https://hunterxhunter.fandom.com/wiki/List_of_Hunter_%C3%97_Hunter_Characters")
    soup = BeautifulSoup(r.content, "html5lib")
    soup.prettify()


    for i in range(488):
        store.append(soup.find_all("img", alt=re.compile("Portrait.png"))[i]["data-src"])

    return store
```
