```python
from bs4 import BeautifulSoup
import requests
import re
```

```python
def getData(link):
  r = requests.get(link)
  soup = BeautifulSoup(r.content, "html5lib")
  soup.prettify()
  comments = soup.find_all("img", alt=re.compile("Portrait.png"))
  return comments
```
