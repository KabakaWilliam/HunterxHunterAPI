```python
from bs4 import BeautifulSoup
import requests
import re
```

```python
def getData(link):
  r = requests.get(link)
  soup = BeautifulSoup(r.content, "html5lib")
  soup.prettify()
  comments = soup.find_all("img", alt=re.compile("Portrait.png"))
  return comments
```

```python
def getName(semiCleanedName):
    flag = True
    flag2="_"
    count=0
    semiCleanedName= semiCleanedName[::-1]

    if semiCleanedName[count] == flag2:
        count+=1
        while flag:
            if semiCleanedName[count] != flag2:
                count+=1
            else:
                flag=False
    semiCleanedName = semiCleanedName[::-1]
    semiCleanedName = semiCleanedName[:-count]
    semiCleanedName = semiCleanedName.replace("_", " ")
    return semiCleanedName
```

```python
for i in range(len(semiCleanedName)):
    semiCleanedName[i] = getName(semiCleanedName[i])
print(semiCleanedName)
    
```
